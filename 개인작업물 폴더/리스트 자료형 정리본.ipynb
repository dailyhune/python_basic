```python
# 리스트 인덱싱 (요소에 접근하기)
a = [1,2,3]  # 리스트 a 생성
a            # 리스트 자체 출력 -> [1, 2, 3]
a[0]         # 인덱스 0번 요소 출력 -> 1
a[0] + a[2]  # 0번 요소와 2번 요소를 더함 -> 1 + 3 = 4
a[-1]        # 마지막 요소 접근 (인덱스 -1) -> 3

# 리스트 안에 리스트가 있는 경우 (중첩 리스트)
a = [1,2,3,['a','b','c']]  # 네 번째 요소가 리스트인 리스트 생성
a[0]         # 첫 번째 요소 출력 -> 1
a[-1]        # 마지막 요소 출력 -> ['a', 'b', 'c']
a[3]         # 인덱스 3의 요소 출력 (위와 동일) -> ['a', 'b', 'c']
a[-1][0]     # 마지막 요소(리스트)에서 첫 번째 값 접근 -> 'a'
a[-1][1]     # 마지막 요소에서 두 번째 값 접근 -> 'b'
a[-1][2]     # 마지막 요소에서 세 번째 값 접근 -> 'c'

# 삼중 리스트에서 인덱싱하기
a = [1,2,['a','b',['Life','is']]]  # 세 번째 요소에 또 다른 리스트가 있는 구조
a[2][2][0]     # 세 번째 요소 → 세 번째 내부 리스트 → 첫 번째 값 -> 'Life'

# 리스트 슬라이싱 (일부 요소 추출)
a = [1,2,3,4,5]
a[0:2]        # 인덱스 0부터 2 전까지의 요소 -> [1, 2]

# 문자열도 슬라이싱 가능
a = "12345"
a[0:2]        # 인덱스 0부터 2 전까지 문자 -> '12'

# 리스트를 나누어 저장
a = [1,2,3,4,5]
b = a[:2]     # 처음부터 인덱스 2 전까지 -> [1, 2]
c = a[2:]     # 인덱스 2부터 끝까지 -> [3, 4, 5]
b             # 결과 확인 -> [1, 2]
c             # 결과 확인 -> [3, 4, 5]

# 중첩 리스트에서 슬라이싱하기
a = [1,2,3,['a','b','c'],4,5]
a[2:5]        # 인덱스 2부터 5 전까지 -> [3, ['a', 'b', 'c'], 4]
a[3][:2]      # 인덱스 3은 리스트 → 그 리스트에서 앞 2개 요소 -> ['a', 'b']

# 리스트 연산
# 리스트 더하기 (+)
a = [1,2,3]
b = [4,5,6]
a + b         # 두 리스트를 이어붙임 -> [1, 2, 3, 4, 5, 6]

# 리스트 반복하기 (*)
a = [1,2,3]
a * 3         # 리스트를 3번 반복 -> [1, 2, 3, 1, 2, 3, 1, 2, 3]

# 리스트의 길이 구하기
a = [1,2,3]
len(a)        # 리스트의 요소 개수 -> 3

# 리스트 연산에서 자주 발생하는 오류
a = [1,2,3]
a[2] + 'hi'   # 숫자와 문자열은 더할 수 없음 → 오류 발생
# TypeError: unsupported operand type(s) for +: 'int' and 'str'

# 숫자를 문자열로 변환 후 문자열 더하기
str(a[2]) + 'hi'  # '3' + 'hi' → '3hi'

# 리스트의 값 수정
a = [1,2,3]
a[2] = 4      # 인덱스 2의 값을 4로 변경
a             # 결과 -> [1, 2, 4]

# 리스트 요소 삭제 - del
a = [1,2,3]
del a[1]      # 인덱스 1의 요소 삭제
a             # 결과 -> [1, 3]

a = [1,2,3,4,5]
del a[2:]     # 인덱스 2부터 끝까지 삭제
a             # 결과 -> [1, 2]

# 리스트 관련 함수들

# 요소 추가 - append
a = [1,2,3]
a.append(4)   # 리스트 끝에 4 추가
a             # 결과 -> [1, 2, 3, 4]
a.append([5,6])  # 리스트 끝에 [5,6]이라는 리스트를 통째로 추가
a             # 결과 -> [1, 2, 3, 4, [5, 6]]

# 리스트 정렬 - sort
a = [1,4,3,2]
a.sort()      # 오름차순 정렬
a             # 결과 -> [1, 2, 3, 4]

a = ['a','c','b']
a.sort()      # 알파벳 순서 정렬
a             # 결과 -> ['a', 'b', 'c']

# 리스트 뒤집기 - reverse
a = ['a','c','b']
a.reverse()   # 순서를 거꾸로 뒤집음
a             # 결과 -> ['b', 'c', 'a']

# 인덱스 반환 - index
a = [1,2,3]
a.index(3)    # 값 3이 있는 인덱스 반환 -> 2
a.index(1)    # 값 1이 있는 인덱스 반환 -> 0

a.index(0)    # 0은 리스트에 없으므로 오류 발생
# ValueError: 0 is not in list

# 요소 삽입 - insert
a = [1,2,3]
a.insert(0,4)   # 인덱스 0에 4 삽입
a               # 결과 -> [4, 1, 2, 3]
a.insert(3,5)   # 인덱스 3에 5 삽입
a               # 결과 -> [4, 1, 2, 5, 3]

# 요소 제거 - remove
a = [1,2,3,1,2,3]
a.remove(3)     # 처음 나오는 3만 제거
a               # 결과 -> [1, 2, 1, 2, 3]
a.remove(3)     # 다음으로 나오는 3 제거
a               # 결과 -> [1, 2, 1, 2]

# 요소 꺼내기 - pop
a = [1,2,3]
a.pop()         # 마지막 요소 꺼냄 → 3 반환하고 리스트에서 삭제
a               # 결과 -> [1, 2]

# 특정 위치 요소 꺼내기
a = [1,2,3]
a.pop(1)        # 인덱스 1 요소 꺼냄 → 2 반환
a               # 결과 -> [1, 3]

# 요소 개수 세기 - count
a = [1,2,3]()



```
